```python
#read in the json C:\Users\CraigParker\OneDrive - Wits PHR\PHR PC\WRHI Files\Documents\Desktop\Heat_Project-1\RP2_Json_0610.json
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import sys

#read in the json file
with open('C:/Users/CraigParker/OneDrive - Wits PHR/PHR PC/WRHI Files/Documents/Desktop/Heat_Project-1/RP2_Json_0610.json') as f:
    data = json.load(f)

```

```python
# Re-confirm the label ID for "Wits RHI"
wits_rhi_label_id = None
for label_id, label_name in data.get('labelNames', {}).items():
    if label_name == 'Wits RHI':
        wits_rhi_label_id = label_id
        break

# Check if the label ID is correctly identified
wits_rhi_label_id
```

```python
# Filter out archived studies and count the number of studies in each stage or bucket
stages_count_excluding_archived = {}
for list_item in data.get('lists', []):
    list_id = list_item.get('id')
    list_name = list_item.get('name')
    stages_count_excluding_archived[list_name] = 0
    for card in data.get('cards', []):
        if card.get('idList') == list_id and not card.get('closed', False):
            stages_count_excluding_archived[list_name] += 1

stages_count_excluding_archived


```

```python
# Identify the label ID for "JHB"
jhb_label_id = None
for label_id, label_name in data.get('labelNames', {}).items():
    if label_name == 'JHB':
        jhb_label_id = label_id
        break

# Collect details of the cards with the specified label "JHB"
jhb_cards = []
if jhb_label_id:
    for card in data.get('cards', []):
        if jhb_label_id in card.get('idLabels', []):
            card_details = {
                'name': card.get('name'),
                'description': card.get('desc'),
                'labels': [data['labelNames'][label] for label in card.get('idLabels', [])],
                'due': card.get('due'),
                'closed': card.get('closed', False)
            }
            jhb_cards.append(card_details)

import ace_tools as tools; tools.display_dataframe_to_user(name="Johannesburg (JHB) Cards Details", dataframe=pd.DataFrame(jhb_cards))

jhb_cards


```
